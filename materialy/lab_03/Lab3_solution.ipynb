## 2. SЕ‚owo kluczowe `assert`

Zanim przejdziemy do frameworkГіw testowych, poznajmy podstawД™: sЕ‚owo kluczowe **`assert`**.

`assert` sprawdza, czy podany warunek jest prawdziwy. JeЕ›li nie вЂ” rzuca wyjД…tek `AssertionError`.

```python
assert warunek, "Opcjonalna wiadomoЕ›Д‡ o bЕ‚Д™dzie"
```

To fundament, na ktГіrym opiera siД™ caЕ‚e testowanie w Pythonie.

````python
# Podstawowe uzycie assert
x = 5
assert x == 5, "x powinno wynosic 5"
print("Asercja przeszla: x == 5")

# Co sie stanie, gdy warunek NIE jest spelniony?
try:
    assert x == 10, "x nie wynosi 10!"
except AssertionError as e:
    print(f"Blad asercji zlapany: {e}")

# Praktyczny przyklad -- testowanie prostej funkcji
def dodaj(a, b):
    return a + b

assert dodaj(2, 3) == 5
assert dodaj(-1, 1) == 0
assert dodaj(0, 0) == 0
print("Wszystkie asercje przeszly pomyslnie!")
````

## 3. unittest вЂ” wbudowany framework testowy

**unittest** to moduЕ‚ wbudowany w Pythona (nie trzeba nic instalowaД‡). Jest wzorowany na frameworku JUnit z Javy.

**Kluczowe zasady unittest:**
- Testy organizujemy w **klasy** dziedziczД…ce po `unittest.TestCase`
- KaЕјda metoda testowa **musi zaczynaД‡ siД™ od `test_`** (inaczej nie zostanie wykryta!)
- UЕјywamy specjalnych metod asercji zamiast zwykЕ‚ego `assert`

### Najpopularniejsze asercje unittest:

| Metoda | Sprawdza |
|--------|----------|
| `assertEqual(a, b)` | a == b |
| `assertNotEqual(a, b)` | a != b |
| `assertTrue(x)` | bool(x) is True |
| `assertFalse(x)` | bool(x) is False |
| `assertRaises(exc)` | czy rzucono wyjД…tek |
| `assertIn(a, b)` | a in b |

### Demo: Testowanie klasy Calculator z unittest

PoniЕјej znajdziesz klasД™ `Calculator` oraz jej testy jednostkowe.

**WAЕ»NE:** Testy unittest najlepiej uruchamiaД‡ z terminala (nie z Jupytera), poniewaЕј `unittest.main()` koЕ„czy proces. Dlatego kod klasy i testГіw znajduje siД™ w osobnych plikach:
- `calculator.py` вЂ” klasa, ktГіrД… testujemy
- `test_calculator_unittest.py` вЂ” testy

PoniЕјej pokazujemy zawartoЕ›Д‡ obu plikГіw, a nastД™pnie uruchamiamy testy komendД… terminalowД….

````python
# ========================================
# Zawartosc pliku: calculator.py
# ========================================

class Calculator:
    """Prosty kalkulator z czterema podstawowymi operacjami."""

    def add(self, a, b):
        return a + b

    def subtract(self, a, b):
        return a - b

    def multiply(self, a, b):
        return a * b

    def divide(self, a, b):
        if b == 0:
            raise ValueError("Dzielenie przez zero!")
        return a / b

# Szybki test reczny -- upewnijmy sie, ze klasa dziala
calc = Calculator()
print(f"2 + 3 = {calc.add(2, 3)}")
print(f"10 - 5 = {calc.subtract(10, 5)}")
print(f"3 * 7 = {calc.multiply(3, 7)}")
print(f"10 / 2 = {calc.divide(10, 2)}")
print("\nKlasa Calculator dziala poprawnie!")
````

````python
# ========================================
# Zawartosc pliku: test_calculator_unittest.py
# ========================================
# (ponizej pokazujemy kod -- NIE uruchamiamy go tutaj,
#  bo unittest.main() zakonczy proces Jupytera)

print("""
import unittest
from calculator import Calculator


class TestCalculator(unittest.TestCase):

    def setUp(self):
        \"\"\"Tworzy nowa instancje Calculator przed KAZDYM testem.\"\"\"
        self.calc = Calculator()

    def test_add_positive_numbers(self):
        result = self.calc.add(2, 3)
        self.assertEqual(result, 5)

    def test_add_negative_numbers(self):
        result = self.calc.add(-1, -1)
        self.assertEqual(result, -2)

    def test_subtract(self):
        result = self.calc.subtract(10, 5)
        self.assertEqual(result, 5)

    def test_multiply(self):
        result = self.calc.multiply(3, 7)
        self.assertEqual(result, 21)

    def test_divide(self):
        result = self.calc.divide(10, 2)
        self.assertEqual(result, 5.0)

    def test_divide_by_zero_raises_error(self):
        with self.assertRaises(ValueError):
            self.calc.divide(10, 0)


if __name__ == "__main__":
    unittest.main()
""")
````

### setUp i tearDown

ZwrГіД‡ uwagД™ na metodД™ `setUp` w powyЕјszym kodzie:

- **`setUp`** вЂ” wywoЕ‚ywana **przed kaЕјdym testem**. Przygotowuje Е›rodowisko (tworzy obiekty, otwiera poЕ‚Д…czenia, ustawia dane).
- **`tearDown`** вЂ” wywoЕ‚ywana **po kaЕјdym teЕ›cie**. SprzД…ta (zamyka pliki, usuwa dane tymczasowe, przywraca stan).

DziД™ki `setUp` kaЕјdy test zaczyna od **czystego stanu** вЂ” testy sД… od siebie **niezaleЕјne**.

```
setUp()  -->  test_add()       -->  tearDown()
setUp()  -->  test_subtract()  -->  tearDown()
setUp()  -->  test_multiply()  -->  tearDown()
...
```

KaЕјdy test dostaje **Е›wieЕјy** obiekt `Calculator()` вЂ” nawet jeЕ›li jeden test go zmodyfikuje, nastД™pny nie odczuje tego efektu.

## 4. pytest вЂ” nowoczesny framework testowy

**pytest** to zewnД™trzna biblioteka, ktГіra upraszcza pisanie testГіw. Kluczowe rГіЕјnice vs unittest:

| Cecha | unittest | pytest |
|-------|----------|--------|
| Struktura | Klasy + metody | ZwykЕ‚e funkcje |
| Asercje | `self.assertEqual(a, b)` | `assert a == b` |
| Przygotowanie | `setUp` / `tearDown` | `@pytest.fixture` |
| Instalacja | Wbudowany | `pip install pytest` |
| Komunikaty bЕ‚Д™dГіw | OgГіlne | SzczegГіЕ‚owe (co dokЕ‚adnie siД™ rГіЕјni) |

**GЕ‚Гіwna zaleta pytest:** mniej kodu, prostszy zapis, lepsze komunikaty bЕ‚Д™dГіw.

Instalacja: `pip install pytest`

### Demo: Te same testy Calculator w pytest

PorГіwnaj poniЕјszy kod z wersjД… unittest вЂ” ten sam kalkulator, ale testy sД… krГіtsze i bardziej czytelne.

Plik: `test_calculator_pytest.py`

````python
# ========================================
# Zawartosc pliku: test_calculator_pytest.py
# ========================================

print("""
import pytest
from calculator import Calculator


@pytest.fixture
def calc():
    \"\"\"Fixture zastepujacy setUp -- tworzy Calculator dla kazdego testu.\"\"\"
    return Calculator()


def test_add(calc):
    assert calc.add(2, 3) == 5
    assert calc.add(-1, 1) == 0


def test_subtract(calc):
    assert calc.subtract(10, 5) == 5


def test_multiply(calc):
    assert calc.multiply(3, 7) == 21


def test_divide(calc):
    assert calc.divide(10, 2) == 5.0


def test_divide_by_zero(calc):
    with pytest.raises(ValueError):
        calc.divide(10, 0)
""")
````

### Fixtures вЂ” przygotowanie Е›rodowiska testowego

**Fixture** w pytest to funkcja oznaczona dekoratorem `@pytest.fixture`. ZastД™puje `setUp` z unittest, ale jest bardziej elastyczna:

- MoЕјe byД‡ **wspГіЕ‚dzielona** miД™dzy wieloma plikami testowymi (przez plik `conftest.py`)
- MoЕјe zwracaД‡ dane lub obiekty
- Fixture moЕјe uЕјywaД‡ innych fixtures (kompozycja)

**Jak to dziaЕ‚a?** Pytest widzi, Ејe funkcja testowa przyjmuje argument `calc` вЂ” szuka fixture o tej nazwie i automatycznie go przekazuje:

```python
@pytest.fixture
def calc():                    # <-- definiujemy fixture
    return Calculator()

def test_add(calc):            # <-- pytest automatycznie przekazuje fixture
    assert calc.add(2, 3) == 5
```

````python
# Demonstracja koncepcji fixture -- prosty przyklad w Jupyterze
# (w prawdziwych testach uzywamy dekoratora @pytest.fixture)

def przygotuj_liste_zakupow():
    """Symulacja fixture -- przygotowuje dane testowe."""
    return ["mleko", "chleb", "maslo", "jajka"]

# "Testy" uzywajace przygotowanych danych
dane = przygotuj_liste_zakupow()

assert len(dane) == 4, "Lista powinna miec 4 elementy"
assert "mleko" in dane, "Mleko powinno byc na liscie"
assert "woda" not in dane, "Wody nie powinno byc na liscie"

print(f"Lista zakupow: {dane}")
print(f"Liczba elementow: {len(dane)}")
print("Wszystkie sprawdzenia przeszly!")
````

### Parametryzacja вЂ” jeden test, wiele zestawГіw danych

**`@pytest.mark.parametrize`** pozwala uruchomiД‡ ten sam test z rГіЕјnymi danymi wejЕ›ciowymi. To eliminuje koniecznoЕ›Д‡ pisania wielu prawie identycznych testГіw.

Dekorator przyjmuje dwa argumenty:
1. Nazwy parametrГіw jako string (oddzielone przecinkami)
2. ListД™ krotek z wartoЕ›ciami

```python
@pytest.mark.parametrize("a, b, expected", [
    (2, 3, 5),       # przypadek 1
    (-1, 1, 0),      # przypadek 2
    (0, 0, 0),       # przypadek 3
])
def test_add(a, b, expected):
    calc = Calculator()
    assert calc.add(a, b) == expected
```

Jeden test, ale pytest uruchomi go **3 razy** вЂ” osobno dla kaЕјdego zestawu danych!

````python
# Uruchamiamy TYLKO sparametryzowany test -- zwroc uwage na output
# Kazdy zestaw danych jest pokazany osobno!
!pytest test_calculator_pytest.py::test_add_parametrized -v
````

### Marks вЂ” oznaczanie testГіw

**Marks** (znaczniki) pozwalajД… kontrolowaД‡, ktГіre testy uruchomiД‡ i jak siД™ majД… zachowaД‡:

- **`@pytest.mark.skip(reason="...")`** вЂ” pomija test bezwarunkowo. Przydatne gdy funkcjonalnoЕ›Д‡ jeszcze nie jest zaimplementowana.
- **`@pytest.mark.xfail(reason="...")`** вЂ” oczekuje niepowodzenia. Test moЕјe nie przejЕ›Д‡, a caЕ‚oЕ›Д‡ i tak bД™dzie "zielona". Przydatne dla znanych bugГіw.

```python
@pytest.mark.skip(reason="Funkcja jeszcze nie zaimplementowana")
def test_future_feature():
    assert new_function() == expected

@pytest.mark.xfail(reason="Znany bug #123")
def test_known_bug():
    assert broken_function() == expected
```
